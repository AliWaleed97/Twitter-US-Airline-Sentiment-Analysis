```python
import nltk
import pandas as pd
from sklearn.model_selection import train_test_split
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
from string import punctuation
import re
from langdetect import detect


```

```python
tweets = pd.read_csv('./Tweets.csv')
X = tweets.drop(columns='airline_sentiment')
Y = tweets['airline_sentiment']
```

```python
all_tweets = tweets['text']
```

# Filtering


```python
for idx, tweet in enumerate(all_tweets):
    if 'RT' in tweet :
        del all_tweets[idx]
        
    if len(tweet)<20:
        del all_tweets[idx]
        
```

```python
# remove urls and strip
all_tweets = [re.sub(r'https?:\/\/.*[\r\n]*', '', text).strip() for text in all_tweets]

# remove punctuation
punctuation += "“’…”"
all_tweets = [''.join([char for char in text if char not in punctuation]) for text in all_tweets]

# tokenize tweets
all_tweets_tokenized = [word_tokenize(text) for text in all_tweets]

# stemming
snowball_stemmer = nltk.stem.SnowballStemmer('english')
all_tweets_tokenized = [[snowball_stemmer.stem(word) for word in words] for words in all_tweets_tokenized]

# # remove stopwords
all_tweets_tokenized = [[word for word in words if word not in stopwords.words('english')] for words in all_tweets_tokenized]

X = [' '.join(words) for words in all_tweets_tokenized]
```
